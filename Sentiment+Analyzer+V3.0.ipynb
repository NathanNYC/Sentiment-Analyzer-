```python
import re 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 


#'C:/Users/nsw26/Desktop/Coding projects/trading_project/fit1.csv'

def read_list():
    source=[]
    with open ("C:/Users/nsw26/Desktop/Coding projects/Sentiment Analyzer/DG/DG3Q19.txt","r") as file1:
        for item in file1:
            source=file1.read().split()
    return source

def read_list2():
    source2=[]
    with open ("C:/Users/nsw26/Desktop/Coding projects/Sentiment Analyzer/DG/DG4Q19.txt","r") as file1:
        for item in file1:
            source2=file1.read().split()
    return source2

def read_list3():
    source3=[]
    with open ("C:/Users/nsw26/Desktop/Coding projects/Sentiment Analyzer/DG/DG1Q20.txt","r") as file1:
        for item in file1:
            source3=file1.read().split()
    return source3


def positive_reference():
    positive=[]
    with open("C:/Users/nsw26/Desktop/Coding projects/Sentiment Analyzer/Positive_reference.txt","r") as file2:
        positive=file2.read().split()
        return positive

def main():
    source=read_list()
    source2=read_list2()
    source3=read_list3()
    
    wordList=[]
    wordList2=[]
    wordList3=[]
    
    positive=positive_reference()
    negative=['not','no','isn\'t','isnt','wasn\'t','can\'t','is not','Not','No',
              'Was not','won\'t','shall not']

    for word in source: 
        if word in positive:
            x=source.index(word)
            #y=string[x-1]
            if source[x-1] not in negative:
                wordList.append(word)
    
    
    for word in source2: 
        if word in positive:
            x=source2.index(word)
            #y=string[x-1]
            if source2[x-1] not in negative:
                wordList2.append(word)
                
    for word in source3: 
        if word in positive:
            x=source3.index(word)
            #y=string[x-1]
            if source3[x-1] not in negative:
                wordList3.append(word)

    
    x=len(wordList)
    y=len(source)
    #print(str(round(x/y*100,2))+'% of the document is positive adjectives')
    
    x2=len(wordList2)
    y2=len(source2)
    #print(str(round(x2/y2*100,2))+'% of the document is positive adjectives')
    
    x3=len(wordList3)
    y3=len(source3)
    #print(str(round(x3/y3*100,2))+'% of the document is positive adjectives')
    
    period1=x/y*100
    period1=round(period1,2)
    
    period2=x2/y2*100
    period2=round(period2,2)
    
    period3=x3/y3*100
    period3=round(period3,3)
    
    price1Lag=154.54  #one month after December 5th 2019... Jan 6th 2020  
    price2Lag=171.57  #one month after March 12th 2020... April 13th 2020 
    price3Lag=188.28  #one month after May 28th 2020... substituting June 11th 2020  
    
    price1Lead=159.01  #one month before December 5th 2019...November 5th 2019 
    price2Lead=158.31  #one month before March 12th 2020... February 12th 2020 
    price3Lead=178.43  #one month before May 28th 2020... April 28th 2020  
    
    
    df=pd.DataFrame([period1,period2,period3],columns=['Positive %'])
    df['Lag Prices']=[price1Lag,price2Lag,price3Lag]
    #df['Lead Prices']=[price1Lead,price2Lead,price3Lead]
    df['Date']=['4Q:19','1Q:20','2Q:20']
    
    
    print(df)
    
    #fig=plt.figure()
    #ax=plt.axes()
    
    fig, ax1=plt.subplots()
    
    color= 'tab:red'
    color2= 'tab:green'
    ax1.set_xlabel('Dates')
    #ax1.xticks(1,2,3,4,5)
    ax1.set_ylabel('DG Share Price ($)')
    ax1.plot(df['Lag Prices'],'o',color=color)
    #ax1.plot(df['Lead Prices'],'o',color=color2)
    
    ax2=ax1.twinx()
    
    ax2.set_ylabel('Positive adjectives %')
    ax2.plot(df['Positive %'],'o')
    
    xlist=(1,2,3,4,5)
    plt.xticks(xlist)
    
    #plt.plot(df['Lag Prices'],'o')
    #plt.plot(df['Lead Prices'],'o')
    #plt.plot(df['Positive %'],'o')
    
    
    plt.show()
    
    
if __name__=='__main__': main()
```
